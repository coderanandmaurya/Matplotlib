## Create Labels for a Plot
With Pyplot, you can use the xlabel() and ylabel() functions to set a label for the x- and y-axis.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([1,2,3,4])
y = x**2

plt.plot(x, y)

plt.xlabel("Number")
plt.ylabel("Square")

plt.show()
```

## **Title** 

```python
plt.plot(x, y)

plt.title("MY first plot")
plt.xlabel("Number")
plt.ylabel("Square")

plt.show()
```

## **Set Font Properties for Title and Labels**

```python
font1 = {'family':'serif','color':'blue','size':25}
font2 = {'family':'serif','color':'darkred','size':15}

plt.title("My first plot", fontdict = font1)
plt.xlabel("Number", fontdict = font2)
plt.ylabel("square", fontdict = font2)

plt.plot(x, y)
plt.show()
```

## **Position the Title**

```python
font1 = {'family':'serif','color':'blue','size':25}
font2 = {'family':'serif','color':'darkred','size':15}

plt.title("My first plot", fontdict = font1,loc = 'left')
plt.xlabel("Number", fontdict = font2)
plt.ylabel("square", fontdict = font2)

plt.plot(x, y)
plt.show()
```

## **Add Grid Lines to a Plot**
With Pyplot, you can use the grid() function to add grid lines to the plot.

```python
plt.grid()
```

```python
plt.grid(axis = 'x')
```

```python
plt.grid(axis = 'y')
```

## **Set Line Properties for the Grid**

```python

x = np.array([1,2,3,4])
y = x**2

plt.plot(x, y)

plt.xlabel("Number")
plt.ylabel("Square")
plt.grid(color = 'green', linestyle = '--', linewidth = 0.5)
plt.show()
```

## **Subplot**
With the subplot() function you can draw multiple plots in one figure

Note: plt.subplot(1, 2, 1)
### the figure has 1 row, 2 columns, and this plot is the first plot.

```python
import matplotlib.pyplot as plt
import numpy as np

#plot 1:
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(1, 2, 1)
plt.plot(x,y)

#plot 2:
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(1, 2, 2)
plt.plot(x,y)

plt.show()
```

## Super Title

```python
import matplotlib.pyplot as plt
import numpy as np

#plot 1:
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(1, 2, 1)
plt.plot(x,y)
plt.title("SALES")

#plot 2:
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(1, 2, 2)
plt.plot(x,y)
plt.title("INCOME")

plt.suptitle("MY SHOP")
plt.show()
```
